```python
def to_snake(name):
    return (
      name.strip()
          .lower()
          .replace("%", "pct") 
          .replace(" or ", "_")
          .replace("  ", " ")
          .replace(" ", "_")
    )

fcc50_df.columns=[to_snake(d) for d in fcc50_df.columns]
```

```python
import pandas as pd

# Load datasets
fcc = pd.read_csv('../data/processed/1939-radio-stations.csv')
apportionment = pd.read_csv('../data/original/apportionment.csv')

# Filter apportionment dataset for Year 1950 and Geography Type 'State'
ap_1940 = apportionment[(apportionment['Year'] == 1940) & (apportionment['Geography Type'].str.lower() == 'state')]

# Standardize state naming: strip whitespace and lowercase
fcc['state_key'] = fcc['state'].str.strip().str.lower()
ap_1940['state_key'] = ap_1940['Name'].str.strip().str.lower()

# Merge on state_key
merged = pd.merge(fcc, ap_1940[['state_key', 'Resident Population']], on='state_key', how='left')

# Drop helper column and rename for clarity
merged = merged.drop(columns=['state_key']).rename(columns={'Resident Population': 'Population_1940'})

# Save merged file for further use
merged.to_csv('../data/processed/merged_fcc_census_1940.csv', index=False)

```
